# Climbing Success Rates of Mount Rainier over a Three Month Period

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime as dt
from sklearn.linear_model import LinearRegression
```

```python
stats = pd.read_csv('climbing_statistics.csv')
stats.Date = pd.to_datetime(stats.Date)
stats['date'] = stats['Date'].map(dt.datetime.toordinal)
```

```python
stats.Route = stats.Route.str.replace(' - ','-')
stats.Route = stats.Route.str.replace(' ','_')
stats.Route = stats.Route.str.replace('\'','')
```

```python
routeUnique = stats.Route.unique()
routeDict = {name : pd.DataFrame() for name in routeUnique}
d = {}

for key in routeDict.keys():
    d[key] = stats[stats.Route == key]
    pd.DataFrame.from_dict(d[key])
    
for key in d:
    d[key] = d[key].reset_index(drop=True)
    key = d[key]
```

```python
sns.pairplot(d['Disappointment_Cleaver'])
```

### Looking at the data as a whole

# Plotting 2014 data(below)

```python
data = d['Disappointment_Cleaver']
name = data.Route.unique()
da = data.Date
time = (da > '2014-07-08') & (da < '2014-10-10')

horiz = da[time]
vert = data['Success Percentage'][time]

#Creating a regression model for 2014
X = data.iloc[:, 5][time].values.reshape(-1, 1)  
Y = data.iloc[:, 4][time].values.reshape(-1, 1)  
linear = LinearRegression()  
linear.fit(X, Y)  
Y_pred = linear.predict(X)  

plt.scatter(horiz, vert)
plt.title(f'Success Rate for Climbing {name} Over Three Months in 2014')
plt.xlabel('Year-Month')
plt.ylabel('Rate of Success')
plt.plot(horiz, Y_pred, color='red')
plt.xticks(rotation = 45)
plt.show()
```

### A regression line of success rates climbing the specified route of mount Rainier in 2014 over the summer months

```python
plt.scatter(horiz, vert)
plt.title(f'Success Rate for Climbing {name} Over Three Months in 2014')
plt.xlabel('Year-Month')
plt.ylabel('Rate of Success')
plt.plot(horiz, Y_pred, color='red')
plt.ylim(Y_pred.min(), Y_pred.max())
plt.xticks(rotation = 45)
plt.show()
```

### A closer look

# Plotting 2015 data(below)

```python
time2 = (da > '2015-7-8') & (da < '2015-10-10')
horiz2 = da[time2]
vert2 = data['Success Percentage'][time2]

#Creating a regression model for 2015
X2 = data.iloc[:, 5][time2].values.reshape(-1, 1)
Y2 = data.iloc[:, 4][time2].values.reshape(-1, 1)
linear.fit(X2, Y2)
Y_pred2 = linear.predict(X2)

plt.scatter(horiz2, vert2)
plt.title(f'Success Rate for Climbing {name} Over Three Months in 2015')
plt.xlabel('Year-Month')
plt.ylabel('Rate of Success')
plt.plot(horiz2, Y_pred2, color='red')
plt.xticks(rotation = 45)
```

### A regression line of success rates climbing the specified route of mount Rainier in 2015 over the summer months

```python
plt.scatter(horiz2, vert2)
plt.title(f'Success Rate for Climbing {name} Over Three Months in 2015')
plt.xlabel('Year-Month')
plt.ylabel('Rate of Success')
plt.plot(horiz2, Y_pred2, color='red')
plt.ylim(Y_pred2.min(), Y_pred2.max())
plt.xticks(rotation = 45)
plt.show()
```

### A closer look

### It looks like earlier in the July is a better time to climb with a little over 50% success rate base on the 2014 and 2015 data. The mean success rates are below

```python
print(f'2014 mean success rate of climbing Mt. Rainier via {name}: ', round(vert.mean(),2))
print(f'2015 mean success rate of climbing Mt. Rainier via {name}: ', round(vert2.mean(),2))
```
