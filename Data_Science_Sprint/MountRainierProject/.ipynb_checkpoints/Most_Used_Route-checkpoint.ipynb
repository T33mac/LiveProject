# The Most Used Climbing Routes on Mt. Rainier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('climbing_statistics.csv')
```

```python
num_used = data
num_used['Total Climbs'] = num_used.Attempted + num_used.Succeeded
```

```python
routes = num_used.groupby(['Route']).sum()
trail_most = routes.sort_values(['Total Climbs'], ascending=False)
trail_most['Route'] = trail_most.index
trail_most = trail_most.reset_index(drop=True)
trail_most = trail_most[trail_most.index < 10]
```

```python
X = trail_most.Route
y = trail_most['Total Climbs']

sns.barplot(X, y)
plt.title('Top 10 Most Used Routes on Mt. Rainier')
plt.xticks(rotation="vertical")
plt.show()
```

### 'Disappointment Cleaver' is the most used route on Rainier
