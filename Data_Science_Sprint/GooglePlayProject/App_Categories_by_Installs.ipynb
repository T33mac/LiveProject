# Visualizing the Popularity of App Categories by the Number of Installs

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
app = pd.read_csv('googleplaystore.csv')
apps = app.drop(['App','Reviews', 'Size', 'Type', 'Price','Rating','Content Rating','Genres', 'Last Updated',
            'Current Ver', 'Android Ver'], axis=1)
apps = apps.drop(apps.index[10472])#This had a '19' rating, causing scale issues
apps.Installs = apps.Installs.str.strip('+')
apps.Installs = apps.Installs.str.replace(',','')#Remove commas so values can be converted to numeric
```

```python
apps['Installs'] = pd.to_numeric(apps['Installs'])
```

### List of categories and number of installs

```python
cat = apps.groupby('Category').agg('sum')
cats = pd.DataFrame(columns=['Category', 'Installs'], data=cat)
cats.Category = cats.index
cats = cats.reset_index(drop=True)
cats
```

### List of 10 most popular app categories

```python
cats = cats.sort_values(by=['Installs'],ascending=False).reset_index(drop=True)
top10 = cats[0:10]
bottom10 = cats[23:33]
top10
```

### List of 10 least popular app categories

```python
bottom10
```

```python
plt.plot(top10.Category, top10.Installs/10e8) #year on x, gdp on y: will show rate of growth
plt.title('Ten Most Popular Categories in Installed Apps (by Billions)', color = 'blue')
plt.xlabel('Category')
plt.xticks(rotation = "vertical")
plt.ylabel('Number of Installs in Billions', color = 'blue')
plt.show()

plt.plot(bottom10.Category, bottom10.Installs/10e5, color ="red") #year on x, gdp on y: will show rate of growth
plt.title('Ten Least Popular Categories in Installed Apps (by Millions)', color = 'red')
plt.xlabel('Category')
plt.xticks(rotation = "vertical")
plt.ylabel('Number of Installs in Millions', color = 'red')
plt.show()
```

### *Game and Communication app categories each top 30 billion installs with a steep drop to the third most popular category, Productivity, at just under 15 billion installs.*

### *Parenting, Beauty, and Events app categories have less than 50 million installs each. The least popular being the Events Category at right around 16 million installs.*
